### Run GSEA analysis for all DEG lists

```r
go_list<-list()
pathway_list<-list()

for(i in 1:length(files)){
    #read in file
    deg = read.csv(file = paste0(indir, files[[i]]))
    names = file_names[[i]]
    #organise data frame
    #trasnlate symbol to gene ID
    for(x in 1:nrow(deg)){
        position = match(deg$X[[x]], gene.df$SYMBOL)[1]
        deg$ENTREZID[x]<-gene.df$ENTREZID[position]
    }

    #filter gene ID and log2FC columns and perform descending rank
    geneList<- deg[,c("ENTREZID", "avg_log2FC")] %>% 
            filter(!is.na(ENTREZID), !is.na(avg_log2FC)) %>% 
            group_by(ENTREZID) %>%
            arrange(-avg_log2FC) %>% # sort descending (important!)
            tibble::deframe() # convert to named vector
    #run gsea and write result to the list
    go_gsea <-gseGO(geneList = geneList, 
                                 ont = "BP", 
                                 OrgDb = "org.Hs.eg.db", 
                                 minGSSize = 15, 
                                 maxGSSize = 500, 
                                 eps = 0, 
                                 nPermSimple = 10000, 
                                 pvalueCutoff = 0.1,
                                verbose = FALSE,
                                 #seed = TRUE
                              )
    go_res<-go_gsea@result
    colnames(go_res)<-paste0(names, "_", colnames(go_res))
    
      pathway_gsea   <- gsePathway(geneList, 
                                        pvalueCutoff = 0.1,
                                        pAdjustMethod = "BH", 
                                        eps = 0, 
                                        verbose = FALSE)
    pathway_res<-pathway_gsea@result
    colnames(pathway_res)<-paste0(names, "_", colnames(pathway_res))

    go_list[[i]] = go_res
    pathway_list[[i]] = pathway_res
}
```

```r
merge_two <- function(x, y) {
  ifelse(
    is.na(x), y,
    ifelse(is.na(y), x, pmax(x, y, na.rm = TRUE))
  )
}

merge_replicates <- function(df, conditions, id_col = "term_description") {
  # extract all non-ID column names
  col_names <- setdiff(names(df), id_col)
  
  # find condition names (everything after "RepX_")
  conditions <- conditions
  
  for (cond in conditions) {
    reps <- grep(cond, col_names, value = TRUE)
    
    if (length(reps) == 1) {
      # only one replicate, just copy it
      merged <- df[[reps]]
    } else {
      # start with the first replicate
      merged <- df[[reps[1]]]
      # merge iteratively with remaining replicates
      for (r in reps[-1]) {
        merged <- merge_two(merged, df[[r]])
      }
    }
    
    # add merged column back to df
    df[[cond]] <- merged
  }
  
  return(df)
}
```

```r
go_merge_rep<-merge_replicates(df = go_merge, conditions = rep_group)

head(go_merge_rep)
```

```r
key_words<-c("antigen processing", "inflammatory response", "endodermal cell differentiation", "cell fate specification",
            "regulation of leukocyte cell-cell adhesion", "neuroblast proliferation"," nuclear division", "mitotic nuclear division",
            "glial cell differentiation")
```

```r
#trim data frame with terms of interest
#test script using ctrl condition

go_ctrl_select<-go_condition_sep[[1]][grep(paste(key_words, collapse = "|"), 
                                           rownames(go_condition_sep[[1]]),
                                          value = TRUE), 
                                     ]

go_ctrl_select[is.na(go_ctrl_select)] <- 0
```

```r
# ---- Extract condition and guide info ----
condition <- str_extract(colnames(go_ctrl_select), "^[^_]+")
guide     <- str_extract(colnames(go_ctrl_select), "[^_]+$")

# ---- Condition-wise ordering ----
condition_order <- order(condition, guide)
mat_conditionwise <- go_ctrl_select[, condition_order]
condition_sorted <- condition[condition_order]
guide_sorted <- guide[condition_order]

# ---- Define annotation bar ----
ha <- HeatmapAnnotation(
  Condition = condition_sorted,
  col = list(Condition = c("krab" = "#1f77b4", "p300" = "#ff7f0e")),
    annotation_name_side = "right"
    )

# ---- Define gaps between conditions ----
gaps <- which(diff(as.numeric(factor(condition_sorted))) != 0)

ht1 <- Heatmap(mat_conditionwise,
               name = "Control group selected GSEA terms NES value",
               column_title = "Condition-wise order",
               cluster_rows = TRUE,
               cluster_columns = FALSE,
              top_annotation = ha,
              column_split = condition_sorted)

pdim(30,10)
draw(ht1, 
     heatmap_legend_side = "bottom", 
     annotation_legend_side = "bottom",
     merge_legends = TRUE)
```

```r
# ---- Ordering Option 2: Guide-wise interleaving ----
guide_order <- order(guide, condition)
guide_sorted <- guide[guide_order]
condition_sorted <- condition[guide_order]

mat_interleave <- go_ctrl_select[, guide_order]
# ---- Define colors for conditions ----
condition_colors <- c("krab" = "#1f77b4", "p300" = "#ff7f0e")
# color gradient for heatmap
heatmap_col_fun <- colorRamp2(
  c(min(mat_interleave), 0, max(mat_interleave)),
  c("#528B8B", "#FFF5EC", "salmon")  # 3-color viridis scale
)

# ---- Build heatmap ----
ht2 <- Heatmap(
  mat_interleave,
  name = "NES",
    col = heatmap_col_fun, 
  cluster_rows = TRUE,
  cluster_columns = FALSE,
  column_labels = guide_sorted,            # guide names only
  column_names_gp = gpar(
    col = condition_colors[condition_sorted], # color by condition
    fontsize = 10
  ),
  bottom_annotation = HeatmapAnnotation(
    Condition = condition_sorted,
    col = list(Condition = condition_colors),
    annotation_name_side = "right"
  ),
  column_title = "Guide-wise interleaved order"
)

# ---- Draw ----
draw(ht2, heatmap_legend_side = "bottom", annotation_legend_side = "bottom", merge_legends = TRUE)
```

```r
# ---- Interleave columns by guide ----
guide_order <- order(guide, condition)
mat_interleave <- go_ctrl_select[, guide_order]
condition_sorted <- condition[guide_order]
guide_sorted <- guide[guide_order]

# ---- Define colors ----
condition_colors <- c("krab" = "#1f77b4", "p300" = "#ff7f0e")

# heatmap color gradient
heatmap_col_fun <- colorRamp2(
  c(min(mat_interleave), 0, max(mat_interleave)),
  c("#528B8B", "#FFF5EC", "salmon")  # 3-color viridis scale
)

pdim(35,10)
# ---- Build heatmap ----
ht <- Heatmap(
  mat_interleave,
  name = "Enrichment",
  col = heatmap_col_fun,
  cluster_rows = TRUE,
  cluster_columns = FALSE,
  #column_split = guide_sorted,                     # one split per guide
  column_title = NULL,                             # remove global title
  show_column_names = FALSE,                          # remove per-column labels
  bottom_annotation = HeatmapAnnotation(            # bottom annotation for conditions
    Condition = condition_sorted,
    col = list(Condition = condition_colors),
    annotation_name_side = "left"
  ),
  # custom text color for condition-wise color distinction
  column_names_gp = gpar(col = condition_colors[condition_sorted])
)

# ---- Draw with guide labels centered ----
ht_drawn<-draw(
  ht,
  heatmap_legend_side = "left",
  annotation_legend_side = "bottom",
  merge_legends = TRUE,
  padding = unit(c(50, 30, 10, 150), "mm"),
  column_title = "Guide-wise interleaved order"
)


decorate_heatmap_body("Enrichment", {
  n_cols <- length(guide_sorted)
  x_positions <- (1:n_cols - 0.5) / n_cols  # evenly spaced across the heatmap width

  for (g in unique(guide_sorted)) {
    cols <- which(guide_sorted == g)
    if (length(cols) > 0) {
      # Find midpoint between first and last column of this guide
      x_mid <- mean(x_positions[cols])
      grid.text(
        g,
        x = unit(x_mid, "npc"),
        y = unit(-10, "mm"),
          rot = 45,  
        just = "right",
        gp = gpar(fontsize = 12, fontface = "bold")
      )
    }
  }
})
```
